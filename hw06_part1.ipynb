# Вывод EM-алгоритма

## Данo:

$X^l$ - выборка данных, каждый элемент характеризуется $n$ параметров.

$y^k$ - номера кластеров,к которым принадлежит элементы выборки



Выборка $X^l$ случайна, независима, из смеси распределений 
$$p(x) = \sum_y w_y p_y(x)$$
$$\sum_y w_y = 1$$


$$X = R^n$$
$$ p_y(x) = (2 * \pi)^{\frac{-n}{2}} (\sigma_{y1} ... \sigma_{yn}) ^ {-1} exp(\frac{-1}{2}\rho_y^2(x, \mu_y))$$

Центры кластеров: $$\mu_y = (\mu_{y1}, ..., \mu_{yn})$$
Диагональная матрица ковариаций: $$\Sigma_y = diag(\sigma_{y1}, ..., \sigma_{yn})$$

$$\rho_y^2(x, x') = \sum^n_{j=1}\sigma_{yj}^{-2} |f_j(x) - f_j(x')|$$

## Найти:

$$w_y = ?$$
$$\mu_{yj} = ? $$
$$\sigma_{yj} = ? $$

$$ y_i = ? $$

---

## Решение

Функция правдоподобия:
$$ Q(\theta, X^l) = \prod_{i=1}^l \phi(x_i, y_i, \theta) $$

Вместо $Q$ используем $-\ln Q$, так как в нем произведение принимает вид суммы.
$$ \ln Q(\theta, X^l) =  \ln {\prod^l_{i=1} \phi(x_i, y_i, \theta)} = \ln {\prod_{i=1}^l p(x_i)} = \sum^l_{i=1} {\ln p(x_i)} = \sum^l_{i=1} ln(\sum^k_(z=1) {w_z p_z(x_i)})  $$

Апостериорная вероятность того, что обучающий объект x i получен из z-й компоненты смеси:
$$g_{iy} = \frac{w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}$$ для всех y, j (3)

Запишем лагранжиан:
$$ L(\theta, X^l) = \sum^l_{i=1} \ln(\sum^k_{z=1} {w_z p_z(x_i)}) - \lambda (\sum^k_{z=1} {w_z} - 1) $$ (4)


Приравняем нулю производную по $w_y$:
$$ \frac{\delta L}{\delta w_y} = \frac{\delta }{\delta w_y}(\sum^l_{i=1} \ln(\sum^k_{z=1} {w_z p_z(x_i)})) - \frac{\delta }{\delta w_y}(\lambda (\sum^k_{z=1} {w_z} - 1)) = \sum^l_{i=1} \frac{\delta }{\delta w_y}( \ln(\sum^k_{z=1} {w_z p_z(x_i)})) - \lambda = \sum^l_{i=1} {\frac {p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}}) - \lambda $$

Умножим обе части на $w_y$
$$\sum^l_{i=1} {\frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}}) - w_y \lambda $$

просуммируем все k этих равенств:
$$\sum^k_z{ \sum^l_{i=1} \frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}}) - \lambda \sum^k_z{w_z} $$

поменяем местами знаки суммирования по j и по i:
$$\sum^l_{i=1} \sum^k_j{\frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}}) - \lambda \sum^k_z{w_z} = $$
$$\sum^l_{i=1} \frac {\sum^k_j{w_y p_y(x_i)}}{\sum^k_{z=1} {w_z p_z(x_i)}}) - \lambda \sum^k_z{w_z} = $$
Согласно ограничениям $\sum^k_z{w_z} = 1$
$$\sum^l_{i=1} \frac {\sum^k_z{w_y p_y(x_i)}}{\sum^k_{z=1} {w_z p_z(x_i)}}) - \lambda 1 = $$
Сокращаем числитель и знаненатель
$$\sum^l_{i=1} 1 - \lambda 1 = 0 $$
$$\lambda = \sum^l_{i=1} 1$$
$$\lambda = n$$

Умножим обе части (4) на $w_y$
$$\sum^l_{i=1} {\frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}}) = w_y \lambda $$
Подставим $\lambda$ 
$$\sum^l_{i=1} {\frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}}) = w_y n $$
$$w_y = \frac{1}{n}\sum^l_{i=1} {\frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}})}$$
Так как $\frac {w_y p_z(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}} = \sum^l_{i=1} g_{iy}$
$$w_y = \frac{1}{l}{\frac {w_y p_z(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}} = \frac{1}{l}\sum^l_{i=1} g_{iy}$$




Приравняем нулю производную по $\theta_{y}$ (общему вектору всех параметров):
$$ \frac{\delta L}{\delta \theta_{y}} = \frac{\delta}{\delta \theta_{y}}(\sum^l_{i=1} \ln(\sum^k_{z=1} {w_z p_z(x_i)})) - \frac{\delta }{\delta \theta_{y}}(\lambda (\sum^k_{z=1} {w_z} - 1)) = \sum^l_{i=1} \frac{\delta }{\delta \theta_y}( \ln(\sum^k_{z=1} {w_z p_z(x_i)})) = \sum^l_{i=1} {\frac {w_y}{\sum^k_{z=1} {w_z p_z(x_i)}}} \frac{\delta}{\delta \theta_{y}} p(x_i) = \sum^l_{i=1} {\frac {w_y p_y(x_i)}{\sum^k_{z=1} {w_z p_z(x_i)}}} \frac{\delta}{\delta \theta_{y}} \ln p(x_i) = \sum^l_{i=1} g_{iy} \frac{\delta}{\delta \theta_{y}} \ln p(x_i) = \frac{\delta}{\delta \theta_{y}} \sum^l_{i=1} g_{iy} \ln p(x_i) $$ j = 1, ..., k. (5)

Если данное условие максимизировать оно будет соотвествовать необходимому условию максимума в задаче максимизации взвешенного правдоподобия 

$$\theta_{y} := arg max_{\theta} \sum^l_{i=1} g_{iy} \ln \phi(x_i; \theta)$$


Приравняем нулю производную по $\mu_{yj}$:
$$ \frac{\delta L}{\delta \mu_y} = \frac{\delta}{\delta \mu_{yj}}(\sum^l_{i=1} \ln(\sum^k_{z=1} {w_z p_z(x_i)})) - \frac{\delta }{\delta \mu_y}(\lambda (\sum^k_{z=1} {w_z} - 1)) = ... = \sum^l_{i=1} g_{iy} \frac{\delta}{\delta \mu_{yj}} \ln p(x_i) = $$
$$\sum^l_{i=1} g_{iy} \frac{\delta}{\delta \mu_{yj}} \ln ((2 * \pi)^{\frac{-n}{2}} (\sigma_{z1} ... \sigma_{zn}) ^ {-1} e^{(\frac{-1}{2}\sum^n_{j=1} (\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj})))^2} = $$
$$ \sum^l_{i=1} g_{iy} \frac{\delta}{\delta \mu_{yj}} ((2 * \pi)^{\frac{-n}{2}} + (\sigma_{z1} ... \sigma_{zn}) ^ {-1} + (\frac{-1}{2}\sum^n_{j=1}(\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj})))^2) $$
$$ \sum^l_{i=1} g_{iy} \frac{-1}{2} \frac{\delta}{\delta \mu_{yj}} (\sum^n_{j=1}(\sigma_{yj}^{-2} (f_j(x_i) - f_j(\mu_{yj})))^2) $$
$$ \sum^l_{i=1} g_{iy} \frac{-1}{2} \frac{\delta}{\delta \mu_{yj}} (\sum^n_{j=1}(\sigma_{yj}^{-2} \frac{\delta}{\delta \mu_{yj}}(f_j(x_i) - f_j(\mu_{yj})))^2) $$
$$ \sum^l_{i=1} g_{iy} \frac{-1}{2} \frac{\delta}{\delta \mu_{yj}} (\sum^n_{j=1} \sigma_{yj}^{-2}(f_j(x_i) - f_j(\mu_{yj})))^2) = $$
$$ \sum^l_{i=1} 2 \frac{-1}{2} \sigma_{yj}^{-2} g_{iy} ((f_j(x_i) - f_j(\mu_{yj})) \frac{\delta}{\delta \mu_{yj}} (f_j(x_i) - f_j(\mu_{yj})) ) = $$
$$ \sum^l_{i=1} \sigma_{yj}^{-2} g_{iy} ((f_j(x_i) - f_j(\mu_{yj})) (f'_j(\mu_{yj}) ) = $$
$$ \sum^l_{i=1} \sigma_{yj}^{-2} g_{iy} ((f_j(x_i) f'_j(\mu_{yj}) - f_j(\mu_{yj}) f'_j(\mu_{yj})) = $$
$$ \sum^l_{i=1} \sigma_{yj}^{-2} g_{iy} ((f_j(x_i) f'_j(\mu_{yj}) - \sum^l_{i=1} g_{iy} \sigma_{yj}^{-2} f_j(\mu_{yj}) f'_j(\mu_{yj})) = 0 $$
$$ \sum^l_{i=1} \sigma_{yj}^{-2} g_{iy}  f_j(\mu_{yj}) f'_j(\mu_{yj})) = \sum^l_{i=1} g_{iy} \sigma_{yj}^{-2} ((f_j(x_i) f'_j(\mu_{yj}) $$
Сокращаем $\sigma_{yj}^{-2} f'_j(\mu_{yj}))$
$$ \sum^l_{i=1} g_{iy}  f_j(\mu_{yj})) = \sum^l_{i=1} g_{iy} (f_j(x_i)) $$
Так как $w_y = \frac{1}{l}\sum^l_{i=1} g_{iy}$
$$ l w_y f_j(\mu_{yj})) = \sum^l_{i=1} g_{iy} (f_j(x_i) $$
$$ f_j(\mu_{yj}) = \frac{1}{l w_y}\sum^l_{i=1} g_{iy} f_j(x_i) $$
вычисляем значение обратное значению от функции $f$ 
$$ \mu_{yj} = f^{-1}_j(\frac{1}{l w_y}\sum^l_{i=1} g_{iy} f_j(x_i)) $$


Приравняем нулю производную по $\sigma_{yj}$:
$$ \frac{\delta L}{\delta \sigma_{yj}} = \frac{\delta}{\delta \sigma_{yj}}(\sum^l_{i=1} \ln(\sum^k_{z=1} {w_z p_z(x_i)})) - \frac{\delta }{\delta \sigma_{yj}}(\lambda (\sum^k_{z=1} {w_z} - 1)) = ... = \sum^l_{i=1} g_{iy} \frac{\delta}{\delta \sigma_{yj}} \ln p(x_i) = $$ 
$$\sum^l_{i=1} g_{iy} \frac{\delta}{\delta \sigma_{yj}} \ln ((2 * \pi)^{\frac{-n}{2}} (\sigma_{z1} ... \sigma_{zn}) ^ {-1} e^{\frac{-1}{2}\sum^n_{j=1} (\sigma_{yj}^{-2} (f_j(x_i) - f_j(\mu_{yj}))^2}) =  $$
$$ \sum^l_{i=1} g_{iy} \frac{\delta}{\delta \sigma_{yj}} ((2 * \pi)^{\frac{-n}{2}} + (\sigma_{z1} ... \sigma_{zn}) ^ {-1} + (\frac{-1}{2}\sum^n_{j=1}\sigma_{yj}^{-2} (f_j(x_i) - f_j(\mu_{yj}))^2) $$
$$ \sum^l_{i=1} g_{iy} \frac{\delta}{\delta \sigma_{yj}} ((\sigma_{y1} ... \sigma_{yn}) ^ {-1} + (\frac{-1}{2}\sum^n_{j=1}\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj})))^2 $$
$$ \sum^l_{i=1} g_{iy} (-\sigma_{yj}^{-2} + \frac{\delta}{\delta \sigma_{yj}}(\frac{-1}{2}\sum^n_{j=1}(\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_yj))))^2) $$
$$ \sum^l_{i=1} g_{iy} (-\sigma_{yj}^{-2} + (2 \frac{-1}{2} (\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj}))) \frac{\delta}{\delta \sigma_{yj}} (\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj})))) $$
$$ \sum^l_{i=1} g_{iy} (-\sigma_{yj}^{-2} + (\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj}))) \frac{\delta}{\delta \sigma_{yj}} (\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj}))) $$
$$ \sum^l_{i=1} g_{iy} ( -\sigma_{yj}^{-2} + (-1) ((\sigma_{yj}^{-1} (f_j(x_i) - f_j(\mu_{yj}))) \sigma_{yj}^{-2} (f_j(x_i) - f_j(\mu_{yj}))) $$
$$ \sum^l_{i=1} g_{iy} (-\sigma_{yj}^{-2} - \sigma_{yj}^{-1} \sigma_{yj}^{-2} (f_j(x_i) - f_j(\mu_{yj}))^2) $$
$$ \sum^l_{i=1} g_{iy} (-\sigma_{yj}^{-2} - \sigma_{yj}^{-3} (f_j(x_i) - f_j(\mu_{yj}))^2) $$
выносим $-\sigma_{yj} ^ {-3} $
$$ -\sigma_{yj} ^ {-3} \sum^l_{i=1} g_{iy}(\sigma_{yj} ^ {1} - (f_j(x_i) - f_j(\mu_{yj}))^2)) = 0 $$
сокращаем $-\sigma_{yj} ^ {-3} $
$$ \sum^l_{i=1} g_{iy} \sigma_{yj} ^ {1} = \sum^l_{i=1} g_{iy} (f_j(x_i) - f_j(\mu_{yj}))^2 $$
Так как $w_y = \frac{1}{l}\sum^l_{i=1} g_{iy}$
$$ l w_y \sigma_{yj} ^ {1} = \sum^l_{i=1} g_{iy} (f_j(x_i) - f_j(\mu_{yj}))^2 $$
$$ \sigma_{yj} = \frac{2}{l w_y} \sum^l_{i=1} g_{iy} (f_j(x_i) - f_j(\mu_{yj}))^2 $$

Так как $g_{yj}$ определяет апостеорную вероятность того, что обучающий объект $x_i$ получен из $y$-ого кластера, то соответсвенно, чтобы найти соответствующий кластер, требуется максимизировать данную апостеорную вероятность по номеру кластера.
$$ y_i = \arg \max_y (g_{iy}) $$

---
Найдено:

 

$$ w_y = \frac{1}{l}\sum^l_{i=1} g_{iy} $$

 

$$ \mu_{yj} = f^{-1}_j(\frac{1}{l w_y}\sum^l_{i=1} g_{iy} f_j(x_i)) $$

 

$$ \sigma_{yj} = \frac{2}{l w_y} \sum^l_{i=1} g_{iy} (f_j(x_i) - f_j(\mu_{yj}))^2 $$

 

$$ y_i = \arg \max_y (g_{iy}) $$
